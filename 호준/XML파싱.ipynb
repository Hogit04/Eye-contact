```python
import xml.etree.ElementTree as ET
import os
import shutil
```

```python
label_list = {'barricade':0, 'bench':1, 'bollard':2, 'chair':3, 'fire_hydrant':4, 'pole':5, 'power_controller':6, 'table':7, 'traffic_sign':8, 'tree_trunk':9, 'crosswalk':10}

entries = os.listdir('/content/drive/MyDrive/팀오시샬라메 케이디지털/자료/원본/인도보행 영상/바운딩박스')
for entry in entries:
    if '.' not in entry :
        se_entries = os.listdir('/content/drive/MyDrive/팀오시샬라메 케이디지털/자료/원본/인도보행 영상/바운딩박스/'+entry)
        for se in se_entries:
            if '.xml' in se :
                tree = ET.parse('/content/drive/MyDrive/팀오시샬라메 케이디지털/자료/원본/인도보행 영상/바운딩박스/'+entry +'/' + se)
                root = tree.getroot()
                for annotations in root:
                    if annotations.attrib.get('name') == None:
                        pass
                    else :
                        name = str(annotations.attrib.get('name')).split('.')[0]+'.txt'
                        filename = '/content/drive/MyDrive/팀오시샬라메 케이디지털/자료/여호준 연습용/' + name
                        with open(filename, 'w') as f:
                            for image in annotations:
                                if image.attrib.get('label') in label_list:
                                    f.write('{} {} {} {} {} \n'.format(label_list[image.attrib.get('label')], 
                                                                    image.attrib.get('xtl'), 
                                                                    image.attrib.get('ytl'), 
                                                                    image.attrib.get('xbr'), 
                                                                    image.attrib.get('ybr')))
                            f.close()
                            print('Done')
```

```python
label_list = ['barricade', 'bench', 'bollard', 'chair', 'fire_hydrant', 'pole', 'power_controller', 'table', 'traffic_sign', 'tree_trunk', 'crosswalk']

entries = os.listdir('/content/drive/MyDrive/공모전/팀오시샬라메 케이디지털/자료/원본/인도보행 영상/바운딩박스/')
for entry in entries:
    if '.' not in entry :
        se_entries = os.listdir('/content/drive/MyDrive/공모전/팀오시샬라메 케이디지털/자료/원본/인도보행 영상/바운딩박스/'+entry)
        for se in se_entries:
            if '.jpg' in se :
                source = r'/content/drive/MyDrive/공모전/팀오시샬라메 케이디지털/자료/원본/인도보행 영상/바운딩박스/'+entry +'/' + se
                destination = r'se'
                shutil.copyfile(source, destination)
                print('Done')
```
