```python
import xml.etree.ElementTree as ET
import os
import shutil
```

```python
label_list = ['barricade', 'bench', 'bollard', 'chair', 'fire_hydrant', 'pole', 'power_controller', 'table', 'traffic_sign', 'tree_trunk', 'crosswalk']
# path = '/content/drive/MyDrive/공모전/팀오시샬라메 케이디지털/자료/train/labels/'
# path = '/content/drive/MyDrive/공모전/팀오시샬라메 케이디지털/자료/여호준연습용/labels/'

dic = {'barricade':0, 'bench':1, 'bollard':2, 'chair':3, 'fire_hydrant':4, 'pole':5, 'power_controller':6, 'table':7, 'traffic_sign':8, 'tree_trunk':9, 'crosswalk':10}

entries = os.listdir('/content/drive/MyDrive/공모전/팀오시샬라메 케이디지털/자료/원본/인도보행 영상/바운딩박스/')
for entry in entries:
    if '.' not in entry :
        se_entries = os.listdir('/content/drive/MyDrive/공모전/팀오시샬라메 케이디지털/자료/원본/인도보행 영상/바운딩박스/'+entry)
        for se in se_entries:
            if '.xml' in se :
                tree = ET.parse('/content/drive/MyDrive/공모전/팀오시샬라메 케이디지털/자료/원본/인도보행 영상/바운딩박스/'+entry +'/' + se)
                root = tree.getroot()
                for annotations in root:
                    if annotations.attrib.get('name') == None:
                        pass
                    else :
                        name = str(annotations.attrib.get('name')).split('.')[0]
                        height = int(annotations.attrib.get('height'))
                        width = int(annotations.attrib.get('width'))
                        filename = name # path + name
                        with open(filename+'.txt', 'w') as f:
                            for image in annotations:
                                if image.attrib.get('label') in label_list:
                                    x1 = float(image.attrib.get('xtl'))
                                    y1 = float(image.attrib.get('ytl'))
                                    x2 = float(image.attrib.get('xbr'))
                                    y2 = float(image.attrib.get('ybr'))

                                    xcenter = abs((x1 + x2)/2/width)
                                    ycenter = abs((y1 + y2)/2/height)
                                    Height = abs((x2 - x1) /height)
                                    Width = abs((y1 - y2) /width)
                                    label = image.attrib.get('label')
                                    f.write('{} {} {} {} {} \n'.format(dic[image.attrib.get('label')], 
                                                                    xcenter, 
                                                                    ycenter, 
                                                                    Height, 
                                                                    Width ))
                            f.close()
                            print('Done')
```
